# Análise de dados

```python
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime
```

```python
def getSeconds(tempo_inicio, tempo_final):
    tempo1 = datetime.strptime(tempo_inicio, "%Y-%m-%d %H:%M:%S")
    tempo2 = datetime.strptime(tempo_final, "%Y-%m-%d %H:%M:%S")
    diferenca = (tempo2 - tempo1).total_seconds()
    
    return(diferenca)
```

# Vídeo 1

## Numero de pessoas detectadas

```python
df_fraco = pd.read_csv('512_32_video1_dados.csv')
df_medio = pd.read_csv('640_32_video1_dados.csv')
df_forte = pd.read_csv('640_48_video1_dados.csv')

plt.figure(figsize=(10, 10))
plt.plot(df_fraco['num_pessoas_detectadas'], label='Dataset 512 x 32')
plt.plot(df_medio['num_pessoas_detectadas'], label='Dataset 640 x 32')
plt.plot(df_forte['num_pessoas_detectadas'], label='Dataset 640 x 48')
plt.xlabel('Índice')
plt.ylabel('numero_pessoas_detectadas')
plt.title('grafico comparando numero de pessoas detectadas dos datasets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print("no vídeo mostra 4 pessoas, menor média = acurária maior")
print(f'o maior valor de pessoas detectadas no dataset 512 x 32 é {df_fraco["num_pessoas_detectadas"].max()} e a média é {df_fraco["num_pessoas_detectadas"].mean()}')
print(f'o maior valor de pessoas detectadas no dataset 640 x 32 é {df_medio["num_pessoas_detectadas"].max()} e a média é {df_medio["num_pessoas_detectadas"].mean()}')
print(f'o maior valor de pessoas detectadas no dataset 640 x 48 é {df_forte["num_pessoas_detectadas"].max()} e a média é {df_forte["num_pessoas_detectadas"].mean()}')
print('-'*100)
max_colunas = df_fraco['num_pessoas_detectadas'].count()
contador_fraco = df_fraco[df_fraco['num_pessoas_detectadas'] == 4].shape[0]
contador_medio = df_medio[df_medio['num_pessoas_detectadas'] == 4].shape[0]
contador_forte = df_forte[df_forte['num_pessoas_detectadas'] == 4].shape[0]

print(f"porcentagem de acerto no modelo 512 X 32 = {(contador_fraco / max_colunas) * 100}%")
print(f"porcentagem de acerto no modelo 640 X 32 = {(contador_medio / max_colunas) * 100}%")
print(f"porcentagem de acerto no modelo 640 X 48 = {(contador_forte / max_colunas) * 100}%")
```

## FPS Atual

```python

plt.figure(figsize=(10, 6))
plt.plot(df_fraco['fps_atual'], label='Dataset 512 x 32')
plt.plot(df_medio['fps_atual'], label='Dataset 640 x 32')
plt.plot(df_forte['fps_atual'], label='Dataset 640 x 48')
plt.xlabel('Valores')
plt.ylabel('fps_atual')
plt.title('grafico comparando fps atual dos datasets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
print(f"MÉDIA - 512 x 32 = {df_fraco['fps_atual'].mean()}")
print(f"MÉDIA - 640 x 32 = {df_medio['fps_atual'].mean()}")
print(f"MÉDIA - 640 x 48 = {df_forte['fps_atual'].mean()}")
```

## Tempo decorrido

```python
tempo_inicio = df_fraco['tempo_atual'][0]
tempo_final = df_fraco['tempo_atual'].iloc[-1]

print(f'Tempo de execução do vídeo 512 x 32: {getSeconds(tempo_inicio, tempo_final)} segundos')

tempo_inicio = df_medio['tempo_atual'][0]
tempo_final = df_medio['tempo_atual'].iloc[-1]
print(f'Tempo de execução do vídeo 640 x 32: {getSeconds(tempo_inicio, tempo_final)} segundos')

tempo_inicio = df_forte['tempo_atual'][0]
tempo_final = df_forte['tempo_atual'].iloc[-1]
print(f'Tempo de execução do vídeo 640 x 48: {getSeconds(tempo_inicio, tempo_final)} segundos')
```

# Conclusão

## Modelo com maior resolução e mais canais de fato representa uma acurácia maior?
Sim. Podemos concluir que de fato, quanto maior a resolução e o valor de canais mais o modelo irá acertar a quantidade de pessoas presentes no vídeo sem demonstrar uma porcentagem grande de erro.
## Modelos mais complexos demoram mais tempo para executar o código? se sim em quanto?
Sim, modelos mais complexos demoram mais para rodar, a diferença entre o modelo mais complexo para o modelo mais simples pode variar de 5% para até 33% de acordo com nossos testes em 4 vídeos.
## Modelos mais complexos são mais pesados e 'rodam' em framerates mais baixos?
Sim, os modelos mais complexos como comprovados nos testes acimas executam em um nível de framerate mais baixo, exigindo mais poder computacional.
## Se modelos mais complexos são melhores, devo usar-los sempre?
Depende, se o objetivo for o poder computacional for limitado e o 'video' a ser testado for de nível simples, talvez não seja necessário utilizar o modelo mais potente, como comprovado no video4, onde o modelo mediano já conseguiu suprir os 100% de acurácia e conseguiu realizar essa tarefa em cerca de 33$ mais rápido do que o modelo mais potente.
